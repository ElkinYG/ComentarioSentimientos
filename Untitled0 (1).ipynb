```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

url = '/content/drive/MyDrive/ElectivaGeneral/Copia de comments_quatro_english.csv'
df = pd.read_csv(url)

def get_sentiment(Comments):
    analysis = TextBlob(Comments)
    print(analysis)
    print(analysis.sentiment.polarity)
    if analysis.sentiment.polarity > 0:  return 'Positive'
    elif analysis.sentiment.polarity == 0: return 'Neutral'
    else: return 'Negative'
# Aplicar la función de sentimiento a cada comentario
df['sentiment'] = df['Comments'].apply(get_sentiment)
# Contar el número de comentarios por sentimiento
sentiment_counts = df['sentiment'].value_counts()
# Visualizar los resultados
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'blue', 'red'])
plt.title('Distribución de Sentimientos en Comentarios de Redes Sociales')
plt.xlabel('Sentimiento')
plt.ylabel('Cantidad de Comentarios')
plt.show()
```
